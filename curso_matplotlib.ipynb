# Aula 03

## Anotações e diversos gráficos

```python
fig = plt.figure(figsize= (12, 6))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(
    dados['data'], 
    dados['temperatura'],
    color = 'tab:blue')

eixo.grid(True)
eixo.set_title('Temperatura no momento', fontsize=20, pad=10)
eixo.set_ylabel('Temperatura', fontsize = 15)
eixo.set_xlabel('Data', fontsize = 15)
eixo.legend(['Temperatura'], loc='best', fontsize=12)

eixo.axhline(max(dados['temperatura']), color= 'k', linestyle = '--')
eixo.axhline(min(dados['temperatura']), color= 'k', linestyle = '--')
```

```python
fig = plt.figure(figsize= (12, 6))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(
    dados['data'], 
    dados['temperatura'],
    color = 'tab:blue')

eixo.grid(True)
eixo.set_title('Temperatura no momento', fontsize=20, pad=10)
eixo.set_ylabel('Temperatura', fontsize = 15)
eixo.set_xlabel('Data', fontsize = 15)
eixo.legend(['Temperatura'], loc='best', fontsize=12)

x1 = dados['data'][dados['temperatura'].idxmax()]
y1 = max(dados['temperatura'])

eixo.annotate('Máximo', xy=(x1,y1), fontsize = 20)

eixo.axhline(max(dados['temperatura']), color= 'k', linestyle = '--')
eixo.axhline(min(dados['temperatura']), color= 'k', linestyle = '--')
```

```python
fig = plt.figure(figsize= (12, 6))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(
    dados['data'], 
    dados['temperatura'],
    color = 'tab:blue')

eixo.grid(True)
eixo.set_title('Temperatura no momento', fontsize=20, pad=10)
eixo.set_ylabel('Temperatura', fontsize = 15)
eixo.set_xlabel('Data', fontsize = 15)
eixo.legend(['Temperatura'], loc='best', fontsize=12)

x1 = dados['data'][dados['temperatura'].idxmax()]
y1 = max(dados['temperatura'])

x2 = dados['data'][dados['temperatura'].idxmax() - 7000]
y2 = max(dados['temperatura'] - 5)

eixo.annotate(  'Máximo', xy=(x1,y1), fontsize = 20,
                xytext = (x2,y2), arrowprops = dict(facecolor = 'k'))

x1 = dados['data'][dados['temperatura'].idxmin()]
y1 = min(dados['temperatura'])

x2 = dados['data'][dados['temperatura'].idxmin() - 7000]
y2 = min(dados['temperatura'] + 5)

eixo.annotate(  'Mínimo', xy=(x1,y1), fontsize = 20,
                xytext = (x2,y2), arrowprops = dict(facecolor = 'k'))

eixo.axhline(max(dados['temperatura']), color= 'k', linestyle = '--')
eixo.axhline(min(dados['temperatura']), color= 'k', linestyle = '--')
```

```python
temperatura_por_semana = dados.groupby('dia_da_semana')['temperatura'].mean()
nome_dias = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
temperatura_por_semana = temperatura_por_semana[nome_dias]
temperatura_por_semana
```

```python
fig = plt.figure(figsize = (5,4))
eixo = fig.add_axes([0,0,1,1])

indice = range(len(temperatura_por_semana))

eixo.bar(indice, temperatura_por_semana)
eixo.set_title('Temperatura por Semana', fontsize = 15, pad=10)
eixo.set_xlabel('Dias da Semana', fontsize = 12)
eixo.set_ylabel('Temperatura Média', fontsize = 12)
```

```python
fig = plt.figure(figsize = (5,4))
eixo = fig.add_axes([0,0,1,1])

indice = range(len(temperatura_por_semana))

eixo.bar(indice, temperatura_por_semana)
eixo.set_title('Temperatura por Semana', fontsize = 15, pad=10)
eixo.set_xlabel('Dias da Semana', fontsize = 12)
eixo.set_ylabel('Temperatura Média', fontsize = 12)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)
```

```python
fig = plt.figure(figsize = (5,4))
eixo = fig.add_axes([0,0,1,1])

indice = range(len(temperatura_por_semana))
cores = ['black', 'r', 'g', 'b', 'yellow', 'orange', 'magenta']

eixo.bar(indice, temperatura_por_semana, color = cores)
eixo.set_title('Temperatura por Semana', fontsize = 15, pad=10)
eixo.set_xlabel('Dias da Semana', fontsize = 12)
eixo.set_ylabel('Temperatura Média', fontsize = 12)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)
```

```python
fig = plt.figure(figsize = (5,4))
eixo = fig.add_axes([0,0,1,1])

indice = range(len(temperatura_por_semana))
cores = ['black', 'r', 'g', 'b', 'yellow', 'orange', 'magenta']

eixo.bar(indice, temperatura_por_semana, color = cores, edgecolor = 'black')
eixo.set_title('Temperatura por Semana', fontsize = 15, pad=10)
eixo.set_xlabel('Dias da Semana', fontsize = 12)
eixo.set_ylabel('Temperatura Média', fontsize = 12)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)
```

```python
fig = plt.figure(figsize = (5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.pie(temperatura_por_semana, labels = temperatura_por_semana.index)
eixo.set_title('Temperatura por Semana', fontsize = 15, pad=10)
```

```python
fig = plt.figure(figsize = (5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.pie(temperatura_por_semana, labels = temperatura_por_semana.index,
autopct='%.2f%%')
eixo.set_title('Temperatura por Semana', fontsize = 15, pad=10)
```

```python
fig = plt.figure(figsize = (5,4))
eixo = fig.add_axes([0,0,1,1])

explodir = [1, 0, 0, 0, 0, 1, 1]

eixo.pie(temperatura_por_semana, labels = temperatura_por_semana.index,
autopct='%.2f%%', explode = explodir)
eixo.set_title('Temperatura por Semana', fontsize = 15, pad=10)
```

```python
fig = plt.figure(figsize = (5,4))
eixo = fig.add_axes([0,0,1,1])

explodir = [0.1, 0, 0, 0, 0, 0.1, 0.1]

eixo.pie(temperatura_por_semana, labels = temperatura_por_semana.index,
autopct='%.2f%%', explode = explodir)
eixo.set_title('Temperatura por Semana', fontsize = 15, pad=10)
```

```python
fig = plt.figure(figsize = (5,4))
eixo = fig.add_axes([0,0,1,1])

explodir = [0.1, 0, 0, 0, 0, 0.1, 0.1]

eixo.pie(temperatura_por_semana, labels = temperatura_por_semana.index,
autopct='%.2f%%', explode = explodir, shadow = True)
eixo.set_title('Temperatura por Semana', fontsize = 15, pad=10)
```
